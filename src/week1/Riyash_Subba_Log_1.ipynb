### 1) Data Structures

```python
# Lists
numbers = [1, 2, 3, 4]
numbers.append(5)
print(numbers)  # Output: [1, 2, 3, 4, 5]

# Dictionaries
student = {"name": "Riyash", "age": 20}
student["grade"] = "A"
print(student)  # Output: {'name': 'Riyash', 'age': 20, 'grade': 'A'}

# Sets
unique_nums = {1, 2, 2, 3}
print(unique_nums)  # Output: {1, 2, 3}

# Tuples
point = (10, 20)
print(point[0])  # Output: 10
```

### 2) Object-Oriented Programming (OOP)

```python
class Chatbot:
    def __init__(self, name): #initializer
        self._name = name  # Encapsulation
    def greet(self):
        return f"Hi, I'm {self._name}!"

class TutorBot(Chatbot):  # Inheritance
    def teach(self, topic):
        return f"{self._name} is teaching {topic}."

bot = TutorBot("TutorGrok")
print(bot.greet())  # Output: Hi, I'm TutorGrok!
print(bot.teach("Python"))  # Output: TutorGrok is teaching Python.
```

### 3) Asynchronous Programming

```python
import asyncio

async def greet_user(name):
    await asyncio.sleep(5)  # Simulate async task
    return f"Hello, {name}!"

async def main():
    result = await asyncio.gather(
        greet_user("Riyash"),
        greet_user("Fragger")
    )
    print(result)  # Output: ['Hello, Riyash!', 'Hello, Fragger!']

await main()
```

### 4) Python Libraries

```python
# NumPy: Arrays and operations
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr * 2)  # Output: [2 4 6 8]

# Pandas: DataFrames
import pandas as pd
data = {"name": ["Alice", "Bob"], "score": [85, 90]}
df = pd.DataFrame(data)
print(df)

# Matplotlib: Basic plot
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [10, 20, 25])
plt.title("Sample Plot")
plt.show()

# Seaborn: Statistical visualization
import seaborn as sns
sns.scatterplot(x="score", y="name", data=df)
plt.show()
```
